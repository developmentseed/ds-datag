# Clustering

```python
import rasterio as rio
from rasterio.plot import show
from sklearn import cluster
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap
```

```python
def raster_to_1d(dataArray):
    imgxyb = np.empty((dataArray.rio.height, dataArray.rio.width, dataArray.rio.count), dataArray.dtype)
    # merge layers as bands/values
    for index in [1,2,3]:
        imgxyb[:,:,index] = dataArray.sel(band=[index])
    # conver 1D array
    img1d=imgxyb[:,:,:].reshape((imgxyb.shape[0]*imgxyb.shape[1],imgxyb.shape[2]))
    print(img1d.shape)
    return imgxyb, img1d
```

```python
# conver raster to 1D
imgxyb, img1d = raster_to_1d(tiles_array[0])
img1d.shape
```

```python
cl = cluster.KMeans(n_clusters=5) # create an object of the classifier
param = cl.fit(img1d) # train it
img_cl = cl.labels_ # get the labels of the classes
img_cl = img_cl.reshape(imgxyb[:,:,0].shape) # reshape labels to a 3d array (one band only)
```

```python
# Create a custom color map to represent our different 4 classes
cmap = LinearSegmentedColormap.from_list("", ["black","red","green","yellow","purple"])
plt.figure(figsize = (7,7))
plt.imshow(img_cl, cmap=cmap, interpolation='nearest', aspect='auto')
plt.colorbar()
plt.show()
```
